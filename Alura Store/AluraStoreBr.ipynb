```python
import pandas as pd

# Carregar os arquivos CSV
loja1 = pd.read_csv("loja_1.csv")
loja2 = pd.read_csv("loja_2.csv")
loja3 = pd.read_csv("loja_3.csv")
loja4 = pd.read_csv("loja_4.csv")

# Verificar as primeiras linhas de cada arquivo
print(loja1.head())
print(loja2.head())
print(loja3.head())
print(loja4.head())
```

```python
# Concatenar os DataFrames
todas_lojas = pd.concat([loja1, loja2, loja3, loja4])

# Verificar a estrutura do DataFrame combinado
print(todas_lojas.info())  # Informações sobre as colunas
print(todas_lojas.head())  # Visualizar os primeiros registros
```

# 3. Média de Avaliação das Lojas

```python
produtos_vendidos = todas_lojas["Produto"].value_counts()

# Produtos mais vendidos
produtos_vendidos.head(10).plot(kind="barh", color="lightgreen", title="Produtos Mais Vendidos")
plt.xlabel("Quantidade Vendida")
plt.ylabel("Produto")
plt.show()

# Produtos menos vendidos
produtos_vendidos.tail(10).plot(kind="barh", color="salmon", title="Produtos Menos Vendidos")
plt.xlabel("Quantidade Vendida")
plt.ylabel("Produto")
plt.show()

```

# 4. Produtos Mais e Menos Vendidos

```python
media_avaliacao_lojas = todas_lojas.groupby("Local da compra")["Avaliação da compra"].mean()
media_avaliacao_lojas.plot(kind="line", marker="o", color="purple", title="Média de Avaliação por Loja")
plt.xlabel("Loja")
plt.ylabel("Média de Avaliação")
plt.grid(True)
plt.show()

```

# 5. Frete Médio por Loja


```python
frete_medio_loja = todas_lojas.groupby("Local da compra")["Frete"].mean()
frete_medio_loja.plot(kind="barh", color="orange", title="Frete Médio por Loja")
plt.xlabel("Frete Médio")
plt.ylabel("Loja")
plt.show()
```
